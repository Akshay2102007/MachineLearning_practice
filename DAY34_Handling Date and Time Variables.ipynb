# Working with Date

```python
#Converting to datetime datatype
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

1. Extract Year


```python
date['date_year'] = date['date'].dt.year
date.head()
```

2. Extract month

```python
date['date_month_no'] = date['date'].dt.month
date.head()
```

```python
#For name
date['date_month_no'] = date['date'].dt.month_name()
date.head()
```

# Extract Day

```python
date['date_day'] = date['date'].dt.day
date.head()
```

```python
# Day of week
date['date_dow'] = date['date'].dt.dayofweek
date.head()
```

```python
#Day of week - name
date['date_dow_name'] = date['date'].dt.day_name()
date.head()
```

```python
#IS weekend
date['date_is_weekend'] = np.where(date['date_dow_name'].isin(['Sunday','Saturday']),1,0)
date.drop(columns=['product_id','city_id','orders']).head()
```

#Extract Week of the year


```python
date['date_week'] = date['date'].dt.isocalendar().week
date.drop(columns=['product_id','city_id','orders']).head()
```

#Extract Quarter


```python
date['date_quarter'] = date['date'].dt.quarter
date.drop(columns=['product_id','city_id','orders']).head()
```

 Extract Sem.

```python
date['semester']=np.where(date['date_quarter'].isin([1,2]),1,2)
date.drop(columns=['product_id','city_id','orders']).head()
```

Extract Time elapsed between dates


```python
import datetime
today = datetime.datetime.today()
today
```

```python
today-date['date']
```

```python
(today - date['date']).dt.days
```

```python
# Months passed
np.round((today - date['date']) / np.timedelta64(1, 'D') / 30.4375, 0) # Dividing by an average number of days per month instead.
```
